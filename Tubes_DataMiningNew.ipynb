# Machine Learning

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  # Mengimpor LinearRegression dari sklearn.linear_model
from sklearn.metrics import mean_squared_error
```

```python
# Load dataset
PPG_Signal = pd.read_csv('/content/mergerAn.csv')
# Menampilkan informasi tentang dataset
print("Informasi PPG_Signal:")
print(PPG_Signal.info())

# Menampilkan beberapa baris pertama dari dataset
print("\nBeberapa baris pertama dari PPG_Signal :")
print(PPG_Signal.head())
```

```python
X = PPG_Signal.drop(columns=['Mean_NBP_Sys', 'Mean_NBP_Dias'])  # Fitur input, hilangkan kolom target
y_sys = PPG_Signal['Mean_NBP_Sys']  # Target untuk Mean_NBP_Sys
y_dias = PPG_Signal['Mean_NBP_Dias']  # Target untuk Mean_NBP_Dias
```

```python
X_train, X_test, y_train_sys, y_test_sys = train_test_split(X, y_sys, test_size=0.2, random_state=42)
```

```python
from sklearn.impute import SimpleImputer
# Menghapus baris dengan nilai NaN
PPG_Signal.dropna(inplace=True)

# Memisahkan fitur dan target setelah penghapusan nilai yang hilang
X = PPG_Signal.drop(columns=['Mean_NBP_Sys', 'Mean_NBP_Dias'])
y_sys = PPG_Signal['Mean_NBP_Sys']

# Memisahkan data latih dan data uji
X_train, X_test, y_train_sys, y_test_sys = train_test_split(X, y_sys, test_size=0.2, random_state=42)

# Menghapus baris dengan nilai NaN pada data latih
X_train.dropna(inplace=True)
y_train_sys = y_train_sys.loc[X_train.index]  # Memastikan target sesuai dengan baris yang tersisa pada X_train

# Inisialisasi imputer
imputer = SimpleImputer(strategy='mean')

# Transformasi data
X_train_imputed = imputer.fit_transform(X_train)

# Langkah 5: Inisialisasi dan latih model Linear Regression untuk Mean_NBP_Sys
model_sys = LinearRegression()
model_sys.fit(X_train_imputed, y_train_sys)
```

```python
y_pred_sys = model_sys.predict(X_test)
```

```python
from sklearn.metrics import r2_score
# Evaluasi model Mean_NBP_Sys
mse_sys = mean_squared_error(y_test_sys, y_pred_sys)
r2_sys = r2_score(y_test_sys, y_pred_sys)

print("Mean Squared Error (Mean_NBP_Sys):", mse_sys)
print("R^2 Score (Mean_NBP_Sys):", r2_sys)

# Menampilkan parameter model Mean_NBP_Sys
print("Intercept (Mean_NBP_Sys):", model_sys.intercept_)
print("Coefficient (Mean_NBP_Sys):", model_sys.coef_)


# Plot hasil prediksi vs nilai aktual
plt.scatter(y_test_sys, y_pred_sys)
plt.xlabel("Nilai Aktual")
plt.ylabel("Prediksi")
plt.title("Perbandingan Nilai Aktual dan Prediksi (Mean_NBP_Sys)")
plt.show()
```

```python
# Bagi data untuk Mean_NBP_Dias
X_train_dias, X_test_dias, y_train_dias, y_test_dias = train_test_split(X, y_dias, test_size=0.2, random_state=42)

# Inisialisasi model regresi linear untuk Mean_NBP_Dias
model_dias = LinearRegression()

# Latih model untuk Mean_NBP_Dias
model_dias.fit(X_train_dias, y_train_dias)

# Lakukan prediksi untuk Mean_NBP_Dias pada data uji
y_pred_dias = model_dias.predict(X_test_dias)
```

```python
import matplotlib.pyplot as plt

# Evaluasi model Mean_NBP_Dias
mse_dias = mean_squared_error(y_test_dias, y_pred_dias)
r2_dias = r2_score(y_test_dias, y_pred_dias)

print("\nMean Squared Error (Mean_NBP_Dias):", mse_dias)
print("R^2 Score (Mean_NBP_Dias):", r2_dias)

# Menampilkan parameter model Mean_NBP_Dias
print("Intercept (Mean_NBP_Dias):", model_dias.intercept_)
print("Coefficient (Mean_NBP_Dias):", model_dias.coef_)

# Plot hasil prediksi vs nilai aktual
plt.scatter(y_test_dias, y_pred_dias)
plt.xlabel("Nilai Aktual")
plt.ylabel("Prediksi")
plt.title("Perbandingan Nilai Aktual dan Prediksi (Mean_NBP_Dias)")
plt.show()
```
